## 3.2 训练和预测过程

**AI生成提示词：**

```
从零实现ID3决策树算法，要求：
1. 只用math.log2，不用其他复杂库
2. 每一行都加详细中文注释
3. 实现信息熵计算和信息增益计算
4. 在注释中说明节点选择过程
5. 实现递归构建决策树的逻辑
6. 实现预测函数
7. 实现数据集划分功能
8. 实现评估指标计算（查准率、查全率、F1）
```

**AI输出：** 生成了完整的决策树算法实现框架，包括所有核心函数。

**手动优化：** 细化了注释内容，补充了关键步骤说明，优化了边界情况处理。

````python
import math
import random

# ===== 信息熵计算（关键注释） =====
def calculate_entropy(data, label_name='购买'):
    """计算数据集的信息熵 H(D) = -∑(p_k * log2(p_k))
    
    信息熵含义:
    - 熵值越高 → 数据集不确定性越大（两类样本比例接近）
    - 熵值越低 → 数据集不确定性越小（某类样本占主导）
    - 熵=0 → 完全纯节点（所有样本属于同一类）
    """
    
    # 步骤1: 统计每个类别的样本数
    label_counts = {}  # 创建字典统计类别
    for sample in data:  # 遍历每个样本
        current_label = sample[label_name]  # 获取标签
        if current_label not in label_counts:  # 如果新类别
            label_counts[current_label] = 0  # 初始化计数
        label_counts[current_label] += 1  # 计数加1
    
    # 步骤2: 计算每个类别的概率并应用熵公式
    entropy = 0.0  # 初始化熵值
    total = len(data)  # 总样本数
    for count in label_counts.values():  # 遍历每个类别
        p = count / total  # 计算概率 p_k = count/total
        if p > 0:  # 避免log2(0)错误
            entropy -= p * math.log2(p)  # 应用公式: -p_k * log2(p_k)
    
    return entropy

# ===== 信息增益计算（关键注释） =====
def calculate_information_gain(data, feature_name, label_name='购买'):
    """计算特征的信息增益 Gain(D,A) = H(D) - ∑(|D_v|/|D| * H(D_v))
    
    信息增益含义:
    - 信息增益 = 划分后不确定性减少的程度
    - 信息增益越大 → 该特征越能区分不同类别
    - 决策树选择信息增益最大的特征作为节点
    """
    
    # 步骤1: 计算划分前的整体熵 H(D)
    total_entropy = calculate_entropy(data, label_name)  # 调用熵函数
    
    # 步骤2: 按特征值划分数据集
    feature_groups = {}  # 创建字典存储分组
    for sample in data:  # 遍历样本
        value = sample[feature_name]  # 获取特征值
        if value not in feature_groups:  # 如果新值
            feature_groups[value] = []  # 创建列表
        feature_groups[value].append(sample)  # 添加样本到对应组
    
    # 步骤3: 计算条件熵 ∑(|D_v|/|D| * H(D_v))
    weighted_entropy = 0.0  # 初始化加权熵
    total = len(data)  # 总样本数
    for subset in feature_groups.values():  # 遍历每个子集
        weight = len(subset) / total  # 计算权重 |D_v|/|D|
        subset_entropy = calculate_entropy(subset, label_name)  # 计算子集熵 H(D_v)
        weighted_entropy += weight * subset_entropy  # 累加加权熵
    
    # 步骤4: 计算信息增益
    gain = total_entropy - weighted_entropy  # Gain = 整体熵 - 条件熵
    return gain

# ===== 选择最优特征（节点选择过程） =====
def find_best_feature(data, features_list, label_name='购买'):
    """选择信息增益最大的特征作为节点
    
    节点选择逻辑:
    - 遍历所有可用特征
    - 计算每个特征的信息增益
    - 选择信息增益最大的特征
    - 该特征最能区分购买/不购买客户
    """
    best_gain = 0.0  # 初始化最大增益
    best_feature = None  # 初始化最优特征
    
    # 遍历所有特征，计算信息增益
    for feature in features_list:  # 遍历特征列表
        gain = calculate_information_gain(data, feature, label_name)  # 计算增益
        if gain > best_gain:  # 如果增益更大
            best_gain = gain  # 更新最大增益
            best_feature = feature  # 更新最优特征
    
    return best_feature, best_gain

# ===== 决策树训练（ID3递归） =====
def train_decision_tree(data, features_list, label_name='购买'):
    """递归构建决策树
    
    ID3算法核心思想:
    - 每次选择信息增益最大的特征作为节点
    - 按特征值划分数据集
    - 对每个子集递归构建子树
    - 两个终止条件: 纯节点或特征耗尽
    """
    
    # 终止条件1: 纯节点（所有样本同一类）
    labels = [sample[label_name] for sample in data]  # 提取标签
    if len(set(labels)) == 1:  # 如果只有一类
        return labels[0]  # 返回类别作为叶子节点
    
    # 终止条件2: 特征耗尽
    if len(features_list) == 0:  # 如果没有特征
        return max(set(labels), key=labels.count)  # 返回多数类
    
    # 选择最优特征作为节点
    best_feature, _ = find_best_feature(data, features_list, label_name)  # 选择节点
    tree = {best_feature: {}}  # 创建树节点
    
    # 移除已用特征
    remaining_features = [f for f in features_list if f != best_feature]  # 剩余特征
    
    # 按特征值划分并递归
    feature_values = set([sample[best_feature] for sample in data])  # 获取所有值
    for value in feature_values:  # 遍历每个值
        subset = [s for s in data if s[best_feature] == value]  # 划分子集
        tree[best_feature][value] = train_decision_tree(subset, remaining_features, label_name)  # 递归
    
    return tree

# ===== 预测函数 =====
def predict(tree, sample, label_name='购买'):
    """使用决策树预测样本类别
    
    预测逻辑:
    - 递归遍历决策树
    - 根据样本特征值选择分支
    - 直到到达叶子节点
    """
    if not isinstance(tree, dict):  # 如果是叶子节点
        return tree  # 返回类别
    
    feature = next(iter(tree))  # 获取当前特征
    value = sample[feature]  # 获取样本特征值
    
    if value not in tree[feature]:  # 如果值不在树中
        return '否'  # 默认返回否
    
    return predict(tree[feature][value], sample, label_name)  # 递归预测

# ===== 数据集划分 =====
def split_dataset(data, train_ratio=0.7, random_seed=42):
    """划分训练集和测试集
    
    划分策略:
    - 70%训练集 + 30%测试集
    - 随机打乱避免顺序偏差
    - 设置随机种子保证可复现
    """
    random.seed(random_seed)  # 设置随机种子
    shuffled = data.copy()  # 复制数据
    random.shuffle(shuffled)  # 随机打乱
    split_point = int(len(shuffled) * train_ratio)  # 计算划分点
    return shuffled[:split_point], shuffled[split_point:]  # 返回训练集和测试集

# ===== 评估指标计算 =====
def calculate_metrics(predictions, actuals, positive='是'):
    """计算查准率、查全率、F1分数
    
    混淆矩阵:
    - TP(真正例): 预测购买且实际购买
    - FP(假正例): 预测购买但实际不购买
    - FN(假负例): 预测不购买但实际购买
    - TN(真负例): 预测不购买且实际不购买
    
    公式:
    - 查准率 = TP/(TP+FP) → 预测购买的准确度
    - 查全率 = TP/(TP+FN) → 实际购买的覆盖率
    - F1 = 2*查准率*查全率/(查准率+查全率) → 综合评分
    """
    TP = sum(1 for p, a in zip(predictions, actuals) if p == positive and a == positive)  # 真正例
    FP = sum(1 for p, a in zip(predictions, actuals) if p == positive and a != positive)  # 假正例
    FN = sum(1 for p, a in zip(predictions, actuals) if p != positive and a == positive)  # 假负例
    TN = sum(1 for p, a in zip(predictions, actuals) if p != positive and a != positive)  # 真负例
    
    # 查准率 Precision = TP/(TP+FP)
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    
    # 查全率 Recall = TP/(TP+FN)
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    
    # F1分数 = 2*P*R/(P+R)
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0
    
    # 准确率
    accuracy = (TP + TN) / (TP + TN + FP + FN)  
    
    return precision, recall, f1, accuracy, {'TP': TP, 'FP': FP, 'FN': FN, 'TN': TN}

print('✓ 决策树算法实现完成')
````

````python
# ===== 训练和评估 =====
print('\n' + '='*50)
print('训练和评估过程')
print('='*50)

# 1. 划分数据集
train_data, test_data = split_dataset(ecommerce_data, train_ratio=0.7, random_seed=42)
print(f'\n数据集划分:')
print(f'  训练集: {len(train_data)}个样本')
print(f'  测试集: {len(test_data)}个样本')

# 2. 训练决策树
print(f'\n训练决策树:')
tree = train_decision_tree(train_data, features, label)
print(f'  ✓ 决策树构建完成')

# 3. 节点选择过程展示
print(f'\n节点选择（计算各特征信息增益）:')
for feature in features:
    gain = calculate_information_gain(train_data, feature, label)
    print(f'  {feature}: {gain:.4f} bits')
best_feature, best_gain = find_best_feature(train_data, features, label)
print(f'  ★ 根节点: {best_feature} (信息增益={best_gain:.4f})')

# 4. 测试集预测
print(f'\n测试集预测:')
predictions = []
actuals = []
for sample in test_data:
    pred = predict(tree, sample, label)
    actual = sample[label]
    predictions.append(pred)
    actuals.append(actual)
    mark = '✓' if pred == actual else '✗'
    print(f'  预测={pred}, 实际={actual} {mark}')

# 5. 计算评估指标
precision, recall, f1, accuracy, conf_matrix = calculate_metrics(predictions, actuals)
print(f'\n混淆矩阵:')
print(f'  TP={conf_matrix["TP"]} (预测买且实际买)')
print(f'  FP={conf_matrix["FP"]} (预测买但实际不买)')
print(f'  FN={conf_matrix["FN"]} (预测不买但实际买)')
print(f'  TN={conf_matrix["TN"]} (预测不买且实际不买)')

print(f'\n评估指标:')
print(f'  查准率: {precision:.4f} ({precision*100:.1f}%)')
print(f'  查全率: {recall:.4f} ({recall*100:.1f}%)')
print(f'  F1分数: {f1:.4f} ({f1*100:.1f}%)')
print(f'  准确率: {accuracy:.4f} ({accuracy*100:.1f}%)')

print('='*50)
````

### 学习收获

通过本项目，我掌握了：
1. **信息熵的计算原理**：熵值反映数据集的不确定性，熵越高不确定性越大
2. **信息增益的计算过程**：信息增益 = 划分后不确定性减少的程度，用于选择最优特征
3. **ID3决策树的递归构建**：每次选择信息增益最大的特征作为节点，递归构建子树
4. **节点选择过程**：遍历所有特征计算信息增益，选择增益最大的特征
5. **评估指标的含义和应用**：
   - 查准率 = 预测购买的准确度（预测买的人中多少真的买）
   - 查全率 = 实际购买的覆盖率（实际买的人中多少被识别）
   - F1分数 = 综合评分（平衡查准率和查全率）
6. **电商业务应用**：如何将算法应用到实际业务场景，提供营销策略建议
