## 第二步：实现信息熵计算
大模型提示词：帮我用 Python 实现一个计算信息熵的函数，输入是上面的西瓜数据集和标签名称 "好瓜"。代码要极其简单，每一行都加详细注释，只用 math.log2，不要用其他复杂库。

```python
import math
def calculate_entropy(data, label):
    # 统计好瓜和坏瓜的数量
    label_counts = {}
    for watermelon in data:
        is_good = watermelon[label]
        if is_good not in label_counts:
            label_counts[is_good] = 0
        label_counts[is_good] += 1
    # 计算总西瓜数
    total = len(data)
    entropy = 0.0
    # 代入公式计算熵
    for count in label_counts.values():
        probability = count / total
        entropy -= probability * math.log2(probability)
    return entropy
print("整个数据集的熵:", calculate_entropy(watermelon_data, label))
```

## 第三步: 实现信息增益计算（找最能区分好瓜的特征）
大模型提示词：基于上面的 calculate_entropy 函数，实现一个计算信息增益的函数。输入是西瓜数据集、标签名称和特征名称。代码要简单，每一行加注释。

```python
def calculate_information_gain(data, label, feature):
    # 计算划分前的总熵
    total_entropy = calculate_entropy(data, label)
    
    # 按特征的不同取值把西瓜分成几堆
    feature_groups = {}
    for watermelon in data:
        value = watermelon[feature]
        if value not in feature_groups:
            feature_groups[value] = []
        feature_groups[value].append(watermelon)
    
    # 计算划分后的加权平均熵
    weighted_entropy = 0.0
    total = len(data)
    
    for group in feature_groups.values():
        group_entropy = calculate_entropy(group, label)
        # 按每堆西瓜的数量加权
        weighted_entropy += (len(group) / total) * group_entropy
    
    # 计算信息增益
    gain = total_entropy - weighted_entropy
    return gain
print("纹理的信息增益:", calculate_information_gain(watermelon_data, label, "纹理"))
```

## 步骤 4：实现找最优特征的函数
大模型提示词：基于上面的信息增益函数，实现一个找最优特征的函数。输入是西瓜数据集、标签名称和特征列表，返回信息增益最大的特征名称。

```python
def find_best_feature(data, label, features):
    best_gain = 0.0
    best_feature = None
    
    for feature in features:
        gain = calculate_information_gain(data, label, feature)
        if gain > best_gain:
            best_gain = gain
            best_feature = feature
    
    return best_feature
print("最优特征:", find_best_feature(watermelon_data, label, features))
```

## 步骤 5：实现决策树训练函数（核心递归部分）
基于上面的所有函数，实现一个递归训练决策树的函数。输入是西瓜数据集、标签名称和特征列表，返回一个字典格式的挑瓜树。代码要加详细注释，特别是递归终止条件的部分。

```python
def train_decision_tree(data, label, features):
    # 终止条件1：这堆西瓜全是好瓜或全是坏瓜
    labels = [watermelon[label] for watermelon in data]
    if len(set(labels)) == 1:
        return labels[0]
    
    # 终止条件2：没有更多特征可以分堆了
    if len(features) == 0:
        # 返回这堆里数量最多的瓜类
        return max(set(labels), key=labels.count)
    
    # 找最能区分好瓜的特征
    best_feature = find_best_feature(data, label, features)
    # 构建树的根节点
    decision_tree = {best_feature: {}}
    
    # 移除已经用过的特征（避免重复提问）
    remaining_features = [f for f in features if f != best_feature]
    
    # 按最优特征的不同取值递归种小树
    feature_values = set([watermelon[best_feature] for watermelon in data])
    for value in feature_values:
        # 筛选出特征值等于当前值的子堆西瓜
        subset = [w for w in data if w[best_feature] == value]
        # 递归训练子树
        decision_tree[best_feature][value] = train_decision_tree(subset, label, remaining_features)
    
    return decision_tree
watermelon_tree = train_decision_tree(watermelon_data, label, features)
print("训练好的挑瓜树:", watermelon_tree)
```

## 步骤 6：实现预测函数（用决策树挑西瓜）
大模型提示词：实现一个预测函数，输入是训练好的挑瓜树和一个西瓜样本（字典格式），返回 "是" 或 "否" 表示是不是好瓜。代码用递归实现，加详细注释。

```python
def predict(tree, watermelon):
    # 获取当前节点问的问题（特征）
    current_feature = next(iter(tree))
    # 获取这个问题的所有可能答案（分支）
    branches = tree[current_feature]
    # 获取这个西瓜的答案
    watermelon_value = watermelon[current_feature]
    
    # 处理训练集中没有出现过的特征值（简单处理）
    if watermelon_value not in branches:
        return "否"
    
    subtree = branches[watermelon_value]
    
    # 如果子树还是字典，继续往下问
    if isinstance(subtree, dict):
        return predict(subtree, watermelon)
    else:
        # 否则返回最终结论
        return subtree
new_watermelon = {"色泽": "乌黑", "根蒂": "蜷缩", "敲声": "浊响", "纹理": "清晰", "脐部": "凹陷", "触感": "硬滑"}
print("新西瓜的预测结果:", predict(watermelon_tree, new_watermelon))
```

## 步骤7：模型评估
大模型提示词：实现一个评估函数，输入是挑瓜树、西瓜数据集和标签名称，返回模型的准确率。

```python
def evaluate(tree, data, label):
    correct = 0
    for watermelon in data:
        prediction = predict(tree, watermelon)
        if prediction == watermelon[label]:
            correct += 1
    accuracy = correct / len(data)
    return accuracy
print("挑瓜树的准确率:", evaluate(watermelon_tree, watermelon_data, label))
```
